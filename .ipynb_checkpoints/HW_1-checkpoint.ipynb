__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$


__Решение__
 
Заметим, что $f_{4}(x)=f_{3}(x)+(-1)f_{2}(x)+(-1)f_{1}(x)$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$ линейно зависимы.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

__Решение__
 
Заметим, что $f_{4}(x)=f_{3}(x)+2f_{2}(x)+1/2f_{1}(x)$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$ линейно зависимы.

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.


__Решение__

Базис линейного пространства $\mathbb{R}^{3}$ образуют векторы 
$b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.
Тогда 

$x=(2, 3, 5)=(2,0,0)+(0, 3, 0)+(0, 0, 5)=\frac{1}{2}\cdot(0, 0, 10)+1\cdot(2, 0, 0)+3\cdot(0, 1, 0)=\frac{1}{2}b_{1}+b_{2}+3b_{3},$


то есть координатами вектора $x$ в данном базисе являются $\frac{1}{2}$, $1$, $3$.

то есть координатами вектора $x$ в данном базисе являются $3x^{2}$, $-2$, $\frac{2}{x^2}$.

__2.__ Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

$$\left\|x\right\|_{} = \sum_{i}|x_{i}|;$$

Определим манхэттенскую норму векторов:

$\left\|x\right\|_{1} = |4|+|2|+|4|=10$

$\left\|x\right\|_{2} = |12|+|3|+|4|=19$

$$\left\|x\right\|_{} = \sqrt{\sum_{i}|x_{i}|^{2}}.$$

Определим евклидову норму векторов:

$\left\|x\right\|_{1} = \sqrt{4^{2}+2^{2}+4^{2}}=\sqrt{36} = 6$

$\left\|x\right\|_{2} = \sqrt{12^{2}+3^{2}+4^{2}}=\sqrt{169} = 13$

$$cos\varphi = \frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|}.$$

$cos\varphi = \frac{4\cdot12+2\cdot3+4\cdot4}{\sqrt{4^{2}+2^{2}+4^{2}}\sqrt{12^{2}+3^{2}+4^{2}}}=\frac{70}{\sqrt{6084}}\approx 0.89$

__4.__ Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$

```python
import numpy as np
from numpy.linalg import norm
```

```python
a = np.array([1, 0, 0])
b = np.array([0, 0, 1])
print(f'Скалярное произведение a и b:\n{np.dot(a, b)}')
print(f'l1 Манхетовская норма вектора a:\n{norm(a, ord=1)}')
print(f'l2 Манхетовская норма вектора b:\n{norm(b, ord=1)}')
print(f'l1 Евклидова норма вектора а:\n{norm(a, ord=2)}')
print(f'l2 Евклидова норма вектора b:\n{norm(b, ord=2)}')
```

```python
a = np.array([1/np.sqrt(2), -1/np.sqrt(2), 0])
b = np.array([1/np.sqrt(2), 1/np.sqrt(2), 0])
c = np.array([0, 0, 1])
print(f'Скалярное произведение a и b:\n{(a @ b )}')
print(f'Скалярное произведение a и c:\n{(a @ c )}')
print(f'Скалярное произведение b и c:\n{(b @ c )}')
print(f'l1 Манхетовская норма вектора a:\n{norm(a, ord=1)}')
print(f'l2 Манхетовская норма вектора b:\n{norm(b, ord=1)}')
print(f'l3 Манхетовская норма вектора c:\n{norm(c, ord=1)}')
print(f'l1 Евклидова норма вектора а:\n{norm(a, ord=2)}')
print(f'l2 Евклидова норма вектора b:\n{norm(b, ord=2)}')
print(f'l3 Манхетовская норма вектора c:\n{norm(c, ord=1)}')
```

```python
a = np.array([1/2, -1/2, 0])
b = np.array([0, 1/2, 1/2])
c = np.array([0, 0, 1])
print(f'Скалярное произведение a и b:\n{(a @ b )}')
print(f'Скалярное произведение a и c:\n{(a @ c )}')
print(f'Скалярное произведение b и c:\n{(b @ c )}')
print(f'l1 Манхетовская норма вектора a:\n{norm(a, ord=1)}')
print(f'l2 Манхетовская норма вектора b:\n{norm(b, ord=1)}')
print(f'l3 Манхетовская норма вектора c:\n{norm(c, ord=1)}')
print(f'l1 Евклидова норма вектора а:\n{norm(a, ord=2)}')
print(f'l2 Евклидова норма вектора b:\n{norm(b, ord=2)}')
print(f'l3 Манхетовская норма вектора c:\n{norm(c, ord=1)}')
```

```python
a = np.array([1, 0, 0])
b = np.array([0, 1, 0])
c = np.array([0, 0, 1])
print(f'Скалярное произведение a и b:\n{(a @ b )}')
print(f'Скалярное произведение a и c:\n{(a @ c )}')
print(f'Скалярное произведение b и c:\n{(b @ c )}')
print(f'l1 Манхетовская норма вектора a:\n{norm(a, ord=1)}')
print(f'l2 Манхетовская норма вектора b:\n{norm(b, ord=1)}')
print(f'l3 Манхетовская норма вектора c:\n{norm(c, ord=1)}')
print(f'l1 Евклидова норма вектора а:\n{norm(a, ord=2)}')
print(f'l2 Евклидова норма вектора b:\n{norm(b, ord=2)}')
print(f'l3 Манхетовская норма вектора c:\n{norm(c, ord=1)}')
```

вывод: а) ортонормированный, б) ортонормированный в евклидовом, в) только а и с образуют ортонормированный базис, г) стандартный ортонормированный базис
