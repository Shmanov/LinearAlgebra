__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$


__Решение__
 
Заметим, что $f_{4}(x)=f_{3}(x)+(-1)f_{2}(x)+(-1)f_{1}(x)$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$ линейно зависимы.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

__Решение__
 
Заметим, что $f_{4}(x)=f_{3}(x)+2f_{2}(x)+1/2f_{1}(x)$, то есть вектор $f_{4}(x)$ — линейная комбинация векторов $f_{1}(x)$, $f_{2}$ и $f_{3}(x)$, из чего можно сделать вывод, что $f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$ линейно зависимы.

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.


__Решение__

Базис линейного пространства $\mathbb{R}^{3}$ образуют векторы 
$b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.
Тогда 

$x=(2, 3, 5)=(2,0,0)+(0, 3, 0)+(0, 0, 5)=\frac{1}{2}\cdot(0, 0, 10)+1\cdot(2, 0, 0)+3\cdot(0, 1, 0)=\frac{1}{2}b_{1}+b_{2}+3b_{3},$


то есть координатами вектора $x$ в данном базисе являются $\frac{1}{2}$, $1$, $3$.

то есть координатами вектора $x$ в данном базисе являются $3x^{2}$, $-2$, $\frac{2}{x^2}$.
